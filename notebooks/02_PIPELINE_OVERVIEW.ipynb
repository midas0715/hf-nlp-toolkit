```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
result = classifier("I was really impressed with how well this project turned out.")
print(result)
```

```python
ner = pipeline("ner", aggregation_strategy="simple")
text = "Pratham is studying at IIMT College of Engineering in Greater Noida."
result = ner(text)
print(result)
```

```python
zero_shot = pipeline("zero-shot-classification")
text = "The new RTX 4050 handles 4-bit quantized LLMs surprisingly well for a laptop GPU."
candidate_labels = ["technology", "sports", "cooking", "finance"]
result = zero_shot(text, candidate_labels)
print(result)
```

```python
fill_mask = pipeline("fill-mask", model="bert-base-uncased")
text = "Pratham is building a project using Hugging Face [MASK] models."
result = fill_mask(text)
for r in result:
    print(r["token_str"], "-", round(r["score"], 4))
```
