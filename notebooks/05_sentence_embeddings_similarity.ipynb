```python
from sentence_transformers import SentenceTransformer

embed_model = SentenceTransformer("all-MiniLM-L6-v2", device="cuda")
```

```python
sentences = [
    "Understanding transformer models is essential for building modern AI applications.",
    "You need to know how transformers work to build good AI systems.",
    "I had pasta for dinner last night."
]

embeddings = embed_model.encode(sentences, convert_to_tensor=True)
print(embeddings.shape)
```

```python
from sentence_transformers.util import cos_sim

similarity_matrix = cos_sim(embeddings, embeddings)
print(similarity_matrix)
```

```python
harder_pair = [
    "The stock market fell sharply after the interest rate announcement.",
    "Shares dropped significantly following news about the rate hike."
]

harder_embeddings = embed_model.encode(harder_pair, convert_to_tensor=True)
harder_sim = cos_sim(harder_embeddings[0], harder_embeddings[1])
print(harder_sim.item())
```

```python
query = "Shares dropped significantly following news about the rate hike."
candidates = [
    "The stock market fell sharply after the interest rate announcement.",
    "You need to know how transformers work to build good AI systems.",
    "I had pasta for dinner last night.",
    "Investors reacted negatively to the central bank's rate decision."
]

query_emb = embed_model.encode(query, convert_to_tensor=True)
candidate_embs = embed_model.encode(candidates, convert_to_tensor=True)

scores = cos_sim(query_emb, candidate_embs)[0]
ranked = sorted(zip(candidates, scores.tolist()), key=lambda x: x[1], reverse=True)

for sentence, score in ranked:
    print(f"{score:.4f}  {sentence}")
```
